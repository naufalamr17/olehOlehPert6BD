```python
import numpy as np

# Remove elements in x that don't have corresponding elements in y
x = [xi for xi, yi in zip(x, y) if not np.isnan(yi)]

# Remove elements in y that don't have corresponding elements in x
y = [yi for xi, yi in zip(x, y) if not np.isnan(yi)]
```
